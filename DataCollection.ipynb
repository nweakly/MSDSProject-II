```python
# initializing a Ring object
from ring_doorbell import Ring

# add your Ring username below
username='your_Ring_user_name_here' 
# add your Ring password below
password='your_Ring_password_here' 

myring=Ring(username, password)
myring.is_connected
```

```python
# listing all door bells linked to the account
myring.doorbells
```

```python
# E.g., Look at the last 10 doorbell events of any kind
for doorbell in myring.doorbells:
    for event in doorbell.history(limit=10):
        print('ID:        %s' %event['id'])
        print('Kind:      %s' %event['kind'])
        print('Answered:  %s' %event['answered'])
        print('When:      %s' %event['created_at'])
        print('--' * 50) 
```

```python
# Make a list of all events with attributes
events_list=pd.DataFrame()
doorbell=myring.doorbells[0]
for event in doorbell.history(limit=100):
    events_list = events_list.append({'event_id': event['id'], 'kind': event['kind'], 'answered': event['answered'], 'date': event['created_at']}, ignore_index=True)
```

```python
# Optional:
# output a list of events to a file for future use
events_list.to_csv('events_list3.csv', date_format='%Y-%m-%d %H:%M:%S')
```
